# 01 EDA and Baseline

Japan-to-Thailand Travel Demand Direction Forecaster, inspired by Moo Deng 🦛


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/sample_jttd_monthly.csv')
df['date'] = pd.to_datetime(df['month'] + '-01')
df.head()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df['date'], df['japanese_visitors_to_thailand'])
plt.title('Monthly Japanese Visitors to Thailand - Sample Data')
plt.xlabel('Month')
plt.ylabel('Visitors')
plt.xticks(rotation=45)
plt.tight_layout()
```
